```python
import pandas as pd
import numpy as np

# Membuat dataset simulasi
data = {
    "Siswa_ID": range(1, 101),
    "Kecepatan_Internet": np.random.choice(["Lambat", "Sedang", "Cepat"], 100, p=[0.4, 0.4, 0.2]),
    "Jenis_Perangkat": np.random.choice(["Ponsel", "Tablet", "Komputer"], 100, p=[0.6, 0.3, 0.1]),
    "Literasi_Digital": np.random.choice(["Rendah", "Menengah", "Tinggi"], 100, p=[0.5, 0.3, 0.2]),
    "Kemampuan_Belajar": np.random.randint(40, 100, 100)
}
df = pd.DataFrame(data)
print(df.head())
```

# **PENGELOMPOKAN DATA**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Encode variabel kategoris
encoder = LabelEncoder()
df['Kecepatan_Internet'] = encoder.fit_transform(df['Kecepatan_Internet'])
df['Jenis_Perangkat'] = encoder.fit_transform(df['Jenis_Perangkat'])
df['Literasi_Digital'] = encoder.fit_transform(df['Literasi_Digital'])

# K-Means Clustering
model = KMeans(n_clusters=3, random_state=42)
clusters = model.fit_predict(df[['Kecepatan_Internet', 'Jenis_Perangkat', 'Literasi_Digital']])
df['Cluster'] = clusters
print(df.head())
```

```python
from tabulate import tabulate

def hash_table_linear_probing(keys, M):
    hash_table = [None] * M
    probing_count = 0

    for key in keys:
        index = key % M
        steps = 0
        while hash_table[index] is not None:
            index = (index + 1) % M
            steps += 1
        hash_table[index] = key
        probing_count += steps + 1

    avg_probing = probing_count / len(keys)
    table = [["Index"] + list(range(M)), ["Value"] + [hash_table[i] if hash_table[i] is not None else " " for i in range(M)]]

    print(tabulate(table, headers="firstrow", tablefmt="grid"))
    print(f"\nRata-rata jumlah probing: {avg_probing:.2f}")

# Contoh penggunaan
keys = [34, 72, 58, 19, 86, 45, 21, 63]
M = min([x for x in keys if not any(x % i == 0 for i in range(2, x))])
hash_table_linear_probing(keys, M)
```
